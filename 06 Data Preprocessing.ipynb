```python
sents = 'the red dog cat eats dog dog eats food red cat eats'
```

```python
from nltk.tokenize import word_tokenize
```

```python
set(word_tokenize(sents))
```

```python
import pandas as pd
```

```python
sentence_1 = 'It was a good practice for us.'
sentence_2 = 'It was also good to know about it.'
sentence_3 = 'For all of us they have done a good practice.'
```

```python
# Find all the distinct words in the senetences
```

```python
words = word_tokenize(
    sentence_1.lower()) + word_tokenize(
    sentence_2.lower()) + word_tokenize(sentence_3.lower())
```

```python
set(words)
```

```python
df = pd.DataFrame(columns=list(set(words)), index=[1,2,3])
```

```python
df
```

```python
tokens1 = word_tokenize(sentence_1.lower())
tokens2 = word_tokenize(sentence_2.lower())
tokens3 = word_tokenize(sentence_3.lower())
```

```python
tokens1
```

```python
tokens2
```

```python
tokens3
```

```python
counts1 = [tokens1.count(token) for token in df.columns]
```

```python
counts1
```

```python
counts2 = [tokens2.count(token) for token in df.columns]
counts2
```

```python
counts3 = [tokens3.count(token) for token in df.columns]
counts3
```

```python
df.iloc[0,:] = counts1
df.iloc[1,:] = counts2
df.iloc[2,:] = counts3
```

```python
# import the class
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# create the object 
cvt = CountVectorizer()
```

```python
# apply the method 
x = cvt.fit_transform([sentence_1,sentence_2,sentence_3])
```

```python
x.toarray()
```

```python
df = pd.DataFrame(data=x.toarray(), columns=cvt.get_feature_names_out())
```

```python
df
```

#### n-grams

```python
cvt = CountVectorizer(ngram_range=(1,2))
```

```python
x = cvt.fit_transform([sentence_1, sentence_2, sentence_3])
```

```python
cvt.get_feature_names_out()
```

```python
df = pd.DataFrame(data=x.toarray(), columns=cvt.get_feature_names_out())
```

```python
df
```

#### TF*IDF Vectorization

```python
# import the class
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# create the object
tfidf = TfidfVectorizer()
```

```python
# apply the method 
x = tfidf.fit_transform([sentence_1, sentence_2,sentence_3])
```

```python
x.toarray()
```

```python
df = pd.DataFrame(data=x.toarray(), columns=tfidf.get_feature_names_out())
```

```python
df
```
