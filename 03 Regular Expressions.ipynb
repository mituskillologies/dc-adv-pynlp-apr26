#### Using re.search

```python
text = "Call me on 555-1234."
```

```python
match = re.search(r"\d{3}-\d{4}", text)
```

```python
match
```

```python
print("Match found:", match.group())
```

#### Using re.findall()

```python
text = '3 apples, 6 mangoes and 14 pineapples.'
```

```python
match = re.findall(r"\d+", text)
```

```python
match
```

#### Using re.sub

```python
text = "I  love       Python Programming   language."
```

```python
clean_text = re.sub(r"\s+"," ", text)
```

```python
clean_text
```

#### Metacharacters

```python
text = 'The cat sat on the mat attached to hat.'
```

```python
re.findall(r".at",text)
```

```python
re.sub(r".at","*", text)
```
