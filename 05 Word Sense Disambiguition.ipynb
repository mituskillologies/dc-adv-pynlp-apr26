#### WSD

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
```

```python
a1 = lesk(word_tokenize('This device is used to jam the signal.'),'jam')
a1.definition()
```

```python
a2 = lesk(word_tokenize('I was stuck in a traffic jam.'),'jam')
a2.definition()
```
